```python
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
live['status_type'] = lb.fit_transform(live['status_type'])
```

```python
live
```

```python
dis = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(live)
    dis.append(kmeanModel.inertia_)
dis
```

```python
plt.figure(figsize=(8,8))
plt.plot(K, dis, 'bx-')
plt.xlabel('k')
plt.ylabel('Dis')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```
